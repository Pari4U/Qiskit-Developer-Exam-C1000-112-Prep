# Creating a quantum circuit

```python
from qiskit import QuantumCircuit
qc = QuantumCircuit(2,2) # (qbit, classicalbit)
qc.measure_all()
qc.draw()
```

```python
qc.draw('mpl')
```

```python
qc.draw('latex')
```

```python
qc = QuantumCircuit(3,2) # (qbit, classicalbit)
qc.measure([0,1], [0,1])
qc.draw('mpl')
```

## Using QuantumRegister and ClassicalRegister

```python
from qiskit import QuantumRegister, ClassicalRegister
from qiskit import Aer, execute
q = QuantumRegister(2)
c = ClassicalRegister(2)
qc = QuantumCircuit(q, c)
qc.measure([0, 1], [0, 1])
qc.draw('mpl')
```

```python
from qiskit.visualization import plot_histogram
sim = Aer.get_backend('statevector_simulator')
job = execute(qc, sim, shots=1024)
result = job.result()
counts = result.get_counts()
plot_histogram(counts)
```

# Single Qubit gates

## Pauli gates

**Pauli-X gate (NOT gate)**

```python
from qiskit.visualization import plot_state_qsphere
from qiskit.quantum_info import Statevector
qc=QuantumCircuit(1)
qc.x(0)
qc.draw(output="mpl")
```

```python
state = Statevector.from_instruction(qc)
state.draw('latex')
```

```python
plot_state_qsphere(state)
```

```python
from qiskit.visualization import array_to_latex

sim = Aer.get_backend('unitary_simulator')
job = execute(qc, sim)
result = job.result()

array_to_latex(result.get_unitary(qc, decimals=3))
#array_to_latex(result.get_statevector(qc)) # for statevector_simulator
```

## P gate($\theta$)

```python
from math import pi
qc=QuantumCircuit(1)
qc.p(pi,0)

qc.draw(output="mpl")
```

```python
job = execute(qc, sim)
result = job.result()

array_to_latex(result.get_unitary(qc, decimals=3))
```
